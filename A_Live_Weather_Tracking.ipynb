````python
# finding the outlair by IQR method
def count_outlair(df):
  outlairs_count={}
  for col in df.select_dtypes(include=['int','float']).columns:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    IQR=q3-q1
    lower_bound=q1-1.5*IQR
    upper_bound=q3+1.5*IQR
    outlairs=df[(df[col]<lower_bound)| (df[col]>upper_bound)]
    outlairs_count[col]=outlairs.shape[0]
  return outlairs_count

print(count_outlair(df))
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd

# Load your dataset
df = pd.read_csv('weather_data.csv')  

# 1. Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 2. Remove duplicates
df = df.drop_duplicates()

# 3. Check & Handle Missing Values
# View missing value count
print("Missing values:\n", df.isnull().sum())

# Fill numeric columns with median
num_cols = ['temprature', 'feels_like', 'temp_min', 'temp_max', 
            'pressure', 'humidity', 'visibility', 'wind_speed']
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical columns with mode
df['city'] = df['city'].fillna(df['city'].mode()[0])
df['weather_main'] = df['weather_main'].fillna(df['weather_main'].mode()[0])
df['weather_description'] = df['weather_description'].fillna(df['weather_description'].mode()[0])
df['country'] = df['country'].fillna(df['country'].mode()[0])

# 4. Convert Time Columns to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df['sunset'] = pd.to_datetime(df['sunset'], errors='coerce')
df['sunrise'] = pd.to_datetime(df['sunrise'], errors='coerce')

# 5. Create Additional Time Features (optional)
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month

# 6. Basic Statistics Summary
print("\nSummary statistics:")
print(df.describe())

# 7. Preview Cleaned Data
print("\nCleaned Data Preview:")
print(df.head())
````

### What all manipulations have you done and insights you found?


- Cleaned and standardized column names
- Removed duplicate rows
- Handled missing values (median for numerics, mode for categoricals)
- Converted `timestamp`, `sunrise`, and `sunset` to datetime
- Created `hour`, `day`, `month` features from `timestamp`
- Generated summary statistics using `.describe()`

---

- Most common weather: **Clear** and **Clouds**
- Temperature shows clear min–max variation daily
- High humidity often linked with **Rain/Mist**
- Wind speed usually low, with occasional peaks
- Day length (sunset - sunrise) varies slightly by city/date
- Minimal missing data; no major data loss after cleaning

````python
def New_Wind_Speed(wind_speed):
    if pd.isnull(wind_speed):
        return 'Unknown'
    elif 1 < wind_speed <= 3:
        return 'Light Air'
    elif 3 < wind_speed <= 7:
        return 'Light Breeze'
    elif 7 < wind_speed < 12:
        return 'Gentle Breeze'
    else:
        return 'Out of Range' 

# Apply function to 'Wind_Speed' column
df['Wind_Speed_Category'] = df['Wind_Speed'].apply(New_Wind_Speed);
````
